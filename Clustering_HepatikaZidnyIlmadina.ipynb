```python
#2
from sklearn import datasets
from sklearn.cluster import KMeans

dataset = datasets.load_iris()
X = dataset.data[:,:]

km = KMeans(n_clusters=5, random_state=1).fit(X)

print (km)
print ('\n')
print ("iterasi :  ", km.n_iter_)
```

```python
from matplotlib import pyplot as plt

#Scatter Plot (Kmeans library)
print ("Scatter Plot untuk KMeans library")
plt.rcParams['figure.figsize'] = (10, 6)
plt.scatter(X[:,0], X[:,1], c=km.labels_, s=24)
center = km.cluster_centers_
plt.scatter(center[:,0], center[:,1], c='red', s=240, marker='x')
plt.show()
```
